## Prepare Data

```python
display(df_latency_dualshock3)
```

```python
# add columns to data frames
df_latency_dualshock3['device'] = 'Dualshock 3'
df_latency_dualshock4['device'] = 'Dualshock 4'

# only one head per cell
#df_latency_dualshock3.head()
df_latency_dualshock4.head()
#print(df_latency_dualshock4['device'])
```

```python
# concatenate two data frames
df_latency = pd.concat([df_latency_dualshock3, df_latency_dualshock4])

df_latency.head()
```

```python
# rename columns
df_latency = df_latency.rename(columns={'latency':'latency_in_ms'})

df_latency.head()
```

```python
# operations on columns
df_latency['latency_in_ms'] = df_latency['latency_in_ms'] / 1000

df_latency.head()
```

```python
# remove columns
df_latency = df_latency.drop(columns=['delayTime'])

df_latency.head()
```

```python
# select only certain columns

df_latency = df_latency[['latency_in_ms', 'device']]

df_latency.head()
```

```python
# selecting one column

df_latency['latency_in_ms'] # prints only one column
```

```python
df_list = list(df_latency['latency_in_ms']) # we can also convert it to other formats!

df_list[:10] # display the first 10 entries
```

```python
# conditional query

df_latency[df_latency['device'] == 'Dualshock 3']['latency_in_ms']
```

### Working with discrete data

```python
# range (min and max)
print("min mass Adelie:    ", df_penguins[df_penguins['species'] == 'Adelie']['body_mass_g'].min())
print("min mass Gentoo:    ", df_penguins[df_penguins['species'] == 'Gentoo']['body_mass_g'].min())
print("min mass Chinstrap: ", df_penguins[df_penguins['species'] == 'Chinstrap']['body_mass_g'].min())
```

```python
# ... a better way:
for species in df_penguins['species'].unique(): 
    mass_min = df_penguins[df_penguins['species'] == species]['body_mass_g'].min()
    mass_max = df_penguins[df_penguins['species'] == species]['body_mass_g'].max()
    
    output = "Mass of {} is between {} g and {} g".format(species, mass_min, mass_max)
    print(output)
```

```python
# calculating the mean
df_penguins[df_penguins['species'] == 'Adelie']['body_mass_g'].mean()
```
